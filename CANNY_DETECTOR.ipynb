##METODE CANNY DETECTOR

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('bunga.jpg',0)
edges = cv.Canny(img,100,200)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Bunga'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Canny Image'), plt.xticks([]), plt.yticks([])
plt.show()

img = cv.imread('panda.jpeg',0)
edges = cv.Canny(img,100,200)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Panda'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Canny Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

##METODE SOBEL DETECTOR

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('bunga.jpg',0)
laplacian = cv.Laplacian(img,cv.CV_64F)
sobelx = cv.Sobel(img,cv.CV_64F,1,0,ksize=5)
sobely = cv.Sobel(img,cv.CV_64F,0,1,ksize=5)
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original Bunga'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show()

img = cv.imread('panda.jpeg',0)
laplacian = cv.Laplacian(img,cv.CV_64F)
sobelx = cv.Sobel(img,cv.CV_64F,1,0,ksize=5)
sobely = cv.Sobel(img,cv.CV_64F,0,1,ksize=5)
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original Panda'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show()
```

#Perbandingan

Metode Sobel menggunakan dua buah kernel dalam implementasinya. Dua kernel tersebut digunakan untuk memperkirakan gradien yang ada pada arah x dan juga pada arah y.
Nilai gradien tersebut dihitung setiap titik atau piksel pada gambar kemudian memberi arah bagi
kernel untuk terus bergerak dan meningkatkan intensitas daricitratersebut


Canny Edge Detector merupakan sebuah metode untuk mendeteksi garis atau tepi dari sebuah citra.

Kriteria meningkatkan deteksi tepi:
1. Tingkat kesalahan yang rendah atau low error rate. Setiap tepi pada citra harus
terdeteksi sebagai tepi sehingga bentuk tidak berubah.
2. Lokalisasi yang baik. Setiap jarak piksel pada tepi yang ditemukan oleh metode Canny
Edge Detector dengan tepi yang sebenarnya harus minimal.
3. Frekuensi tingkat tepi yang terdeteksi secara ganda harus rendah. Ini bertujuan agar
tidak menimbulkan kemungkinan yang lebih dari citrayangdideteksi.

##Hasil
Edge detector menggunakan model canny menghasilkan detaksi garis tepi yang lebih jelas terlihat , sedangkan metode sobel menghasilkan deteksi tepi seperti tekstur sehingga kurang terlihat jelas tepinya.



##kesimpulan
Penggunaan metode Canny Edge Detector dan Sobel Edge Detector memanfaatkan ciri berupa bentuk dan tekstur dalam melakukan prosesidentifikasi, namun
metode canny lebih baik saat digunakan untuk mendeteksi tepi dari citra gambar.
